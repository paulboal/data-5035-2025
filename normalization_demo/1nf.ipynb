# 1-NF Normalization using Pandas Melt

```python
import pandas as pd
```

```python
raw = pd.read_excel('Normalization Examples.xlsx', sheet_name='0-NF')
```

```python
raw
```

```python
# Melt down for each Course Grade
#  "<Course> Grade"
#  Grade
nf = raw.melt(
    id_vars=raw.columns[0:4],
    value_vars=raw.columns[range(4,17,2)],
    var_name='Course',
    value_name='Grade'
)
```

```python
nf
```

```python
# Strip the word "Grade" from the Course name
nf['Course'] = nf['Course'].apply(lambda x: x.split(' ')[0])
```

```python
nf
```

```python
# Do a second melt to get the Instructor
nf['Instructor'] = raw.melt(
    id_vars=raw.columns[0:4],
    value_vars=raw.columns[range(5,18,2)],
    value_name='Instructor'
)['Instructor']
```

```python
nf
```

```python
raw.columns=list(raw.columns[0:4]) + [x.split(' ')[1] + '-' + x.split(' ')[0] for x in raw.columns[4:]]
```

```python
raw['id'] = raw.index
```

```python
raw
```

```python
pd.wide_to_long(
    raw,
    stubnames=['Instructor','Grade'],
    i='id',
    j='Course',
    sep='-',
    suffix='\w+'
)
```
